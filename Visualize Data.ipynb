### Biểu diễn mối liên hệ giữa giá sản phẩm và rating của người dùng

```python
rating_info = card_info[['price', 'rating', 'rating_num']]
rating_info
```

```python
# Check sự tương quan giữa giá và rating
rating_info[['price', 'rating']].corr()
```

```python
# Check sự phân bổ số lượng rating
fig = plt.subplots(figsize =(10, 7))
plt.hist(rating_info['rating_num'], bins = 20)
```

##### Có thể chia các sản phẩm thành 4 nhóm dựa theo số lượng rating
- Nhóm 1: không có rating
- Nhóm 2: < 50 ratings
- Nhóm 3: 50 - 100 ratings
- Nhóm 4: > 100 ratings

```python
# Function để phân loại card theo số lượng ratings
def categorise(row):  
    if row['rating_num'] == 0:
        return 'No ratings'
    elif row['rating_num'] < 50:
        return '<50 ratings'
    elif row['rating_num'] >= 50  and row['rating_num'] <= 100:
        return '50 - 100 ratings'
    return '>100 ratings'
```

```python
# Thêm cột phân loại sản phẩm theo số lượng ratings
rating_info['rating_type'] = rating_info.apply(lambda row: categorise(row), axis=1)
rating_info
```

```python
rating_info.groupby('rating_type')['rating_type'].size()
```

```python
# Các sản phẩm No rating
card_no_rating = rating_info.loc[rating_info['rating_type'] == 'No ratings', 'price']
card_no_rating
```

```python
# Sự phân bổ giá của các sản phẩm không có rating
fig = plt.subplots(figsize =(10, 7))
plt.hist(card_no_rating, bins = 10)
```

```python
# Các sản phẩm có ratings
fig = plt.subplots(figsize =(10, 7))
sns.scatterplot(x="price", y="rating", hue="rating_type", s=100, data=rating_info, alpha = 0.5);
```

#### Nhận xét:
- Giá và rating của các loại card không có sự tương quan với nhau
- Có gần 1700 sản phẩm không có rating, các sản phẩm này đa số có mức giá < 1000 đô
- Các sản phẩm < 750 đô có số lượng đánh giá tương đối lớn, mức đánh giá dao động chủ yếu từ 4.0-5.0
